```python
#Count the number of flights for each airline and display the results in a bar graph, sorted in ascending order
data['airline'].value_counts(ascending=True).plot.barh(color =['lightgreen','lightblue'])
plt.title("Airlines with frequencies")
plt.xlabel("No of Flight")
plt.ylabel("Airline")
plt.show()
```

```python
#showing the departure time and arrival time for flight with their count          
plt.figure(figsize=(16,4))
plt.subplot(1, 2, 1)
plt.bar(data['departure_time'].value_counts().index ,data['departure_time'].value_counts().values, color =['r','b'] )
plt.title("Department Time")
plt.xlabel("D. Time")
plt.ylabel("D. Freq")
plt.subplot(1, 2, 2)

plt.bar(data['arrival_time'].value_counts().index ,data['arrival_time'].value_counts().values, color =['g','y'] )
plt.title("Arrival Time")
plt.xlabel("A. Time")
plt.ylabel("A. Freq")

plt.show()
```
